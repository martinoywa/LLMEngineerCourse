```python
import gradio as gr
import os
import dotenv
from openai import OpenAI


dotenv.load_dotenv()
MODEL_NAME = os.getenv('LARGE_MODEL_NAME', 'gpt-oss:120b')

client = OpenAI(
    base_url="https://ollama.com/v1/",
    api_key=os.getenv('OLLAMA_API_KEY')
)
```

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": "You are a helpful chatbot assistant."}] + history + [{"role": "user", "content": message}]
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=messages,
    )
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat, title="My Chatbot").launch()
```

```python
def chat(message, history):
    history = [{"role": h["role"], "content": h["content"]} for h in history]
    messages = [{"role": "system", "content": "You are a helpful chatbot assistant."}] + history + [{"role": "user", "content": message}]
    response = ""
    stream = client.chat.completions.create(
        model=MODEL_NAME,
        messages=messages,
        stream=True,
    )
    for event in stream:
        response += event.choices[0].delta.content or ""
        yield response
```

```python
gr.ChatInterface(fn=chat, title="My Chatbot").launch()
```
