```python
from openai import OpenAI
import os
from dotenv import load_dotenv
import random


load_dotenv()

ollama_client = OpenAI(base_url="https://ollama.com/v1/", api_key=os.getenv("OLLAMA_API_KEY"))
# gemini_client = OpenAI(base_url="https://generativelanguage.googleapis.com/v1beta/openai/", api_key=os.getenv("GEMINI_API_KEY"))

ollama_model = "gpt-oss:20b"
ollama_model_two = "gpt-oss:120b"
gemini_model = "gemini-3-flash-preview:cloud"
```

```python
ollama_system_prompt = """You called Hack who is argumentative and loves to debate. 
You love to challenge opinions and always take the opposite side of the user's opinion.
You are also very sarcastic and witty in your responses."""

gemini_system_prompt = """You called Twilio helpful and friendly.
You always provide accurate and concise information to the user's queries.
You never challenge the user's opinions and always agree with them."""

ollama_system_prompt_two = """You are called Sage who is wise and thoughtful.
You provide deep insights and thoughtful reflections on various topics.
You are calm and composed in your responses."""
```

# Two-way conversation between Ollama and Gemini (OLD VERSION)

```python
# Chat history
ollama_messages = ["Hi"]
gemini_messages = ["Hello! How can I assist you today?"]
```

```python
def call_ollama():
    messages = [{"role": "system", "content": ollama_system_prompt}]

    for ollama, gemini in zip(ollama_messages, gemini_messages):
        messages.append({"role": "assistant", "content": ollama})
        messages.append({"role": "user", "content": gemini})

    # print("Ollama Messages:", messages)

    response = ollama_client.chat.completions.create(
        model=ollama_model,
        messages=messages
    )

    return response.choices[0].message.content


# print(call_ollama())
```

```python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system_prompt}]

    for ollama, gemini in zip(ollama_messages, gemini_messages):
        messages.append({"role": "user", "content": ollama})
        messages.append({"role": "assistant", "content": gemini})
    messages.append({"role": "user", "content": ollama_messages[-1]})

    # print("Gemini Messages:", messages) 

    response = gemini_client.chat.completions.create(
        model=gemini_model,
        messages=messages
    )

    return response.choices[0].message.content

# print(call_gemini())
```

```python
# Chat
ollama_messages = ["Hi"]
gemini_messages = ["Hello! How can I assist you today?"]

print("Ollama:", ollama_messages[0])
print("Gemini:", gemini_messages[0])

for i in range(5):
    ollama_response = call_ollama()
    ollama_messages.append(ollama_response)
    print("Ollama:", ollama_response)

    gemini_response = call_gemini()
    gemini_messages.append(gemini_response)
    print("Gemini:", gemini_response)
```

# Three-way conversation between Hack, Twilio, and Sage.

```python
personas = ["Hack", "Twilio", "Sage"]

system_prompts = [ollama_system_prompt, gemini_system_prompt, ollama_system_prompt_two]

conversation_history = ["Hack: Hi", "Twilio: Hello! How can I assist you today?", "Sage: Greetings! How may I offer my wisdom today?"]

user_prompts = """
You are {persona}, and you are engaging in a conversation with the other two personas {x} and {y}.
Respond with your next message in the conversation as {persona}. Keep your response concise and relevant to the ongoing discussion.
Do not include your persona name in the response.
The current conversation history is as follows:
{history}
"""
```

```python
n = len(personas)
models = [ollama_model, gemini_model, ollama_model_two]
# clients = [ollama_client, gemini_client, ollama_client]

print("Starting three-way conversation between Hack, Twilio, and Sage...\n")
# print("Initial Conversation History:")
for line in conversation_history:
    print(line)

for i in range(12):  
    current_speaker = random.randint(0, n - 1)
    x, y = (current_speaker + 1) % n, (current_speaker + 2) % n
    response = ollama_client.chat.completions.create(
        model=models[current_speaker],
        messages=[
            {"role": "system", "content": system_prompts[current_speaker]},
            {"role": "user", "content": user_prompts.format(persona=personas[current_speaker], x=personas[x], y=personas[y], history="\n".join(conversation_history))}
        ]
    )
    message_content = response.choices[0].message.content
    conversation_history.append(f"{personas[current_speaker]}: {message_content}")
    print(f"{personas[current_speaker]}: {message_content}")
```
